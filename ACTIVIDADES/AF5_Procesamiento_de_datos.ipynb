*Ejercicio 1.- Análisis y normalización de datos de sensores en un robot móvil*

```python
#1.Se importan las librerias que vamos a utilizar
import pandas as pd #importamos pandas
import matplotlib.pyplot as plt #importamos matplotlib
import numpy as np #importamos numpy
from sklearn.preprocessing import MinMaxScaler #importamos MinMaxScaler

#2.Se carga el dataset (punto 1 de la actividad), en este caso lo haremos directo del repositorio
#Otra opción es descargar el archivo y añadir la ruta en donde se guardó
url = "https://raw.githubusercontent.com/dilp90/InteligenciaArtificial_y_RedesNeuronales_UANL_FIME/refs/heads/main/MachineLearning/Datasets/robot_sensors.csv"
df = pd.read_csv(url)

#3.Mostramos las pimeras filas del dataset (punto 2 de la actividad)
print("Primeras filas del dataset:")
print(df.head()) #se observará una columna adicional, es el indice que asigna automaticamente
```

```python
#4.Limpiamos los datos (punto 3 de la actividad)
#Inciso a
df = df.dropna()  # Elimina las filas con valores nulos
#Inciso b, filtrar valores atipicos
df = df[(df["Velocidad (cm/s)"] >= 0) & (df["Distancia (cm)"] >= 0) & (df["Tiempo (s)"] >= 0)]
#en este caso se toma en cuenta velocidad, distancia y tiempo, que deben ser mayor o igual a 0

#5.Normalizar los datos (punto 4 de la actividad)
scaler = MinMaxScaler()
df[["Distancia (cm)", "Velocidad (cm/s)", "Temperatura (°C)"]] = scaler.fit_transform(
    df[["Distancia (cm)", "Velocidad (cm/s)", "Temperatura (°C)"]]
)

#6.Visualizar los datos normalizados (punto 6), vamos a utilizar matplotlib
#Gráfico 1:Distancia vs Tiempo
plt.scatter(df["Tiempo (s)"], df["Distancia (cm)"], label="Distancia", color="blue")
plt.xlabel("Tiempo (s)")
plt.ylabel("Distancia (Normalizada)")
plt.title("Distancia vs Tiempo")
plt.legend()
```

```python
#Gráfico 2: Temperatura vs Tiempo
plt.scatter(df["Tiempo (s)"], df["Temperatura (°C)"], label="Temperatura", color="red")
plt.xlabel("Tiempo (s)")
plt.ylabel("Temperatura (Normalizada)")
plt.title("Temperatura vs Tiempo")
plt.legend()
```

```python
#Gráfico 3: Velocidad vs Tiempo
plt.scatter(df["Tiempo (s)"], df["Velocidad (cm/s)"], label="Velocidad", color="green")
plt.xlabel("Tiempo (s)")
plt.ylabel("Velocidad (Normalizada)")
plt.title("Velocidad vs Tiempo")
plt.legend()
```
